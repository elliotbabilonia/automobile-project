### Automobile Models

This notebook explores various aspects of automobile models, investigating their impact on pricing. Key features compared to price: `highway-mpg`, `body-style`, and `engine-size` were examined. The numerical visualizations were followed with descriptive statistics analyzing the mean, standard deviation, etc. The categorical visualizations observed the differences between price on the different categories we observed such as 'make', and 'body-style'. The analysis contributes valuable insights into the factors influencing the pricing of automobile models.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from ucimlrepo import fetch_ucirepo 
import warnings 
warnings.filterwarnings('ignore')
  
# Fetching the dataset 
automobile = fetch_ucirepo(id=10)
```

```python
automobile.data.features['highway-mpg']
```

```python
highwaympg_counts=automobile.data.features['highway-mpg'].value_counts().reset_index()
highwaympg_counts
```

```python
# Comparing the Highway Miles per Gallon to the Price

sns.histplot(automobile.data.features['highway-mpg'], kde=True)

sns.set_style("whitegrid")
sns.set_palette("pastel")

plt.xlabel('Highway MPG')
plt.ylabel('Price')
plt.title('Highway MPG vs Price')
```

```python
automobile.data.features.describe()['highway-mpg']
```

```python
# Comparing the body-style to the Price

plt.figure(figsize=(5, 5))

sns.boxplot(x=automobile.data.features['body-style'], y=automobile.data.features['price'])

plt.title('body-style vs Price')
plt.show()
```

```python
sns.barplot(x=automobile.data.features["make"], y=automobile.data.features["price"])
plt.xticks(rotation=90)
```

```python
automobile.data.features['engine-size']
```

```python
enginesize_counts=automobile.data.features['engine-size'].value_counts().reset_index()
enginesize_counts
```

```python
# Comparing the Engine Size to the Price

sns.histplot(automobile.data.features['engine-size'], kde=True)

plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.title('Engine Size vs Price')
```

```python
automobile.data.features.describe()['engine-size']
```

```python
# Comparing the Distribution of the Engine Types to the Price

engine_value_counts = automobile.data.features["engine-type"].value_counts()

plt.figure(figsize=(5, 5))
sns.barplot(x=engine_value_counts.index, y=engine_value_counts.values)

plt.xticks(rotation='vertical')
plt.xlabel('Engine Type')
plt.ylabel('Price')
plt.title('Distribution of Engine Types')

plt.show()
```
